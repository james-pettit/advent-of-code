```python
with open('day7_input.txt') as file:
    line = file.readline()

crab_pos = [int(i) for i in line.strip().split(",")]

fuel_required_list = []

for selected_pos in range(max(crab_pos)):
    fuel = 0
    for pos in crab_pos:
        fuel += sum(range(abs(selected_pos-pos)+1))
    fuel_required_list.append(fuel)
    
print(min(fuel_required_list))

import numpy as np 
import matplotlib.pyplot as plt

plt.clf()
plt.plot(fuel_required_list)
plt.rcParams['figure.figsize'] = [8,8]
plt.rcParams['figure.dpi'] = 100
plt.show()
```
